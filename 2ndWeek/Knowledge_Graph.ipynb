**Logic**: The study of how to make formal correct deductions and inferences.

When we talk about ***a Logic***, we mean a formal language, which has

1.   **Symbols**: Alphabet (characters) which can be used to form sequences in the language.
2.   **Syntax**: Rules (grammar) which prescribe how to construct well-formed sequences of symbols (strings).<br>
i.e. *well-formed symbol sequences without meaning*
3. **Semantics**: Rules which define how the meaning of complex sequences of symbols can be derived from atomic sequences of symbols.<br> i.e. *gives meaning to syntax*




**Formal Semantics**: The rules to derive meaning are so mathematically precise and unambiguous, that they can be programmed, so the meaning of well-formed sequences can be derived completely automatically by a machine.

As the goal of Sematics is usually to model some real-world phenomena, it can also be defined in Model Theoretic terms: 

The Semantics of a Logic has an associated Model, and Meaning is defined as a formal Interpretation within the Model.

#### **[Propositional Logic](https://en.wikipedia.org/wiki/Propositional_calculus)** (from Wikipedia):


Propositional Logic deals with **propositions**, which are statements that have a **truth value** either True or False. <br>

**Compound propositions** are formed by connecting **atomic formulas** (propositions) using **logical connectives** (e.g. AND, OR).<br>

The truth value of compound propositions can be calculated from their component parts using truth tables of the connectives (**compositionality**).<br>

Using the truth table method, the truth value of any arbitrary formula can be effectively determined. This means that Propositional Logic is **decidable**.

Propositional Logic also deals with **argumentation**. A set of formulas is an **argument**, iff (if and only if) the last formula in the set (the **conclusion**) can be derived from the previous ones (**premises**) using the **inference rules** in the logic.









image.png

#### **[First Order Logic](https://en.wikipedia.org/wiki/First-order_logic)** (from Wikipedia):

**First Order Logic (FOL)** additionally uses **variables**, **quantification** (universal and existential quantifiers) and relations (**predicates**). <br>These make the language much more **expressive**.<br>

On the other hand, there is a price to be paid for the additional expressiveness in terms of complexity.

While Propositional Logic is decidable, FOL is only **semi-decidable**. There is an effective way to find all valid formulas (**theorems**), but there is no effective way to determine that a formula is **not in the theory**.

Wikipedia's [First Order Logic page](https://en.wikipedia.org/wiki/First-order_logic) is very clear and well structured. Please read at least the [Introduction](https://en.wikipedia.org/wiki/First-order_logic#Introduction)

### Description Logics

**Description Logics (DLs)** are a fragments of FOL, which find the right balance between expressiveness and computational complexity.<br>
DLs 

*   model ***concepts***, ***roles*** and ***individuals*** and their relationships
*   use ***constructors*** to create more complex descriptions from simple descriptions
*   differ from each other (family of DLs) in the constructors they use
* are **decidable** (most times, depending on the constructors)
* were originally developed from "sematic networks"




The following slide from the video shows the structure of a DL Knowledge Base (Ontology):

image.png

Let's take a closer look at the smallest deductively complete DL, **ALC (Attribute Language with Complement)**:

*   Class constructors are **Conjunction**, **Disjunction** and **Complement**
*   Quantifiers restrict domain and range of **Roles**
*   Has **Classes**, **Roles/Properties** and **Individuals**



The model-theoretic semantic for ACL is an **Interpretation**, which consists of a **set of Individuals**, and an **Interpretation Function**.<br>
The Interpretation Function defines the following mappings:

*   each name in the syntax is mapped to an individual
*   a class is mapped to a set of individuals
*   a relation is mapped to pairs of individuals

### Inference problems in a DL Knowledge Base

*   **Global Consistency**<br>
Are there contradictory facts in the KB? In other words, can a contradiction be derived from the facts in the KB?<br>
This is the most important question, since only globally consistent KBs are useful.

*   **Class Consistency**<br>
If a Class is inconsistent, it must necessarily be empty in order for the KB not to become inconsistent.

*   **Class Inclusion (Subsumption)**<br>
Are there classes whose members are necessarily members of an another class as well?

*   **Class Equivalence**<br>
Are there classes whose members are necessarily members of an another class and vice versa? In other words, are there classes which are subclasses of each other?

*   **Class Disjunctiveness**<br>
Are there classes whose intersection is always necessarily empty?
That is, an individual cannot be a member of both classes at the same time?

*   **Class Membership**<br>
Is an individual contained in a particular class?

*   **Instance Generation (Retrieval)**<br>
Find all (known) individuals contained in a class.







Since DLs are fragments of FOL, **FOL inference algorithms can be used for DLs as well**. <br>
The mechanics of these inference algorithms is such, that they can be used to **show unstatisfiability of a theory (KB)**, that is, to find contradictions. <br>So we need to formulate entailment problems (statements we want to prove) in such a way that the **proof by contradiction** technique can be applied.<br>
Let's look again at our inference problems above, and re-formulate them:<br>
(Notation: "*iff*" means "*if and only if*")




### Proof by contradiction

*   **Global Consistency**<br>
Our KB is inconsistent (unsatisfiable), iff a contradiction can be derived from it.

*   **Class Consistency**<br>
A class is inconsistent iff it cannot have any members. In order to use the reduction to unsatisfiability technique, we can reformulate this: <br>
1. Assume that the class has some member, and add this fact to the KB.
2. If this makes the KB unsatisfiable (a contradiction can derived), it means that that our assumption that the class can have a member was wrong, and we proved that it is inconsistent.

*   **Class Inclusion (Subsumption)**<br>
Assume that the subclass has a member that is not included in the superclass and derive a contradiction.

*   **Class Equivalence**<br>
Prove class inclusion (previous point) in both directions.

*   **Class Disjunctiveness**<br>
Assume that the classes have a member in common and derive a contradiction.

*   **Class Membership**<br>
Assume that the particular indivudual is not included in the class, and derive a contradiction.

*   **Instance Generation (Retrieval)**<br>
Check class membership (previous point) for all individuals.

#### Exercise: Proofs in the African Wildlife Ontology

1.   Prove that Lion does not eat Warthog.
2.   Prove that Giraffe is a Herbivore.



*Solution:*

1. Warthogs eat some Animal, so they are not Herbivores. But Lions eat only Herbivores. A contradiction.

2. Assume that Giraffes are not Herbivores. Giraffes eat only Leafs or Twigs. An Animal is a Herbivore if it eats only Plants or proper parts of Plants. Since Leafs and Twigs are both proper parts of Plants, Giraffes satisfy the definition of Herbivore. This contradicts our assumption that Giraffes are not Herbivores.

### The Open World Assumption (OWA)

In an empty Ontology, **anything is possible**.
So Ontologies need to be defined by iteratively constraining what is possible.<br>
In other words, under the **Open World Assumption**, anything can be true what is not explicitly denied -- we only know if something is not true, if assuming it leads to a contradiction.<br>

The most important consequence of the OWA is, that **we can say "don't know"** (based on our knowledge (KB)). <br>
If either a statement or its negation can be added to a KB without making it inconsistent, the statement cannot be decided based on the KB, and it returns "don't know" as an answer.<br>

Also, the existence of further individuals is possible if they are not explicitly excluded. So just because an individual is not explicitly in the KB, it does not mean it does not exist.<br>

These characteristics of the OWA are very important for real life applications, so the OWA is the default assumption for Knowledge Graphs.

**Logic-based vs. Learning-based reasoning**

As we saw above, logic-based reasoning has its roots in mathematical proof-theory, and has a very precise definion.
On the other hand, in the literature about learning-based reasoning, typically no formal definition of reasoning is given, it is treated as a prediction task. 

Let's compare and contrast the two approaches:




|   	|  Logic-based 	|   Learning-based	|
|---	|---	|---	|
|  Advantages	|  <ul><li>accurate answers</li><li>identifies all  inferences</li></ul> | <ul><li>scalable</li><li>robust to noise</li></ul>| 
|  Disadvantages 	|  Difficulties with handling <ul><li>large</li><li>incomplete</li><li>conflicting</li><li>uncertain</li></ul> data. | <ul><li>probabilistic accuracy</li></ul>  	|




#### **Recursive Reasoning Networks (RRN)**

RRNs aim at closing the gap between the two approaches, by trying to combine the advantages of both.<br>
For the formal definition of the problem they are designed to solve, please read Section 2. of the paper.

In informal terms, the task is to decide about any given fact whether it is implied by the facts and rules already in the Knowledge Base.

````
# Imports and definitions

from __future__ import print_function
from functools import reduce
import re
import tarfile

import numpy as np

from keras.utils.data_utils import get_file
from keras.layers.embeddings import Embedding
from keras import layers
from keras.layers import recurrent
from keras.models import Model
from keras.preprocessing.sequence import pad_sequences

import os

def tokenize(sent):
    '''Return the tokens of a sentence including punctuation.

    >>> tokenize('Bob dropped the apple. Where is the apple?')
    ['Bob', 'dropped', 'the', 'apple', '.', 'Where', 'is', 'the', 'apple', '?']
    '''
    return [x.strip() for x in re.split(r'(\W+)', sent) if x.strip()]


def parse_stories(lines, only_supporting=False):
    '''Parse stories provided in the bAbi tasks format

    If only_supporting is true,
    only the sentences that support the answer are kept.
    '''
    data = []
    story = []
    for line in lines:
        line = line.decode('utf-8').strip()
        nid, line = line.split(' ', 1)
        nid = int(nid)
        if nid == 1:
            story = []
        if '\t' in line:
            q, a, supporting = line.split('\t')
            q = tokenize(q)
            if only_supporting:
                # Only select the related substory
                supporting = map(int, supporting.split())
                substory = [story[i - 1] for i in supporting]
            else:
                # Provide all the substories
                substory = [x for x in story if x]
            data.append((substory, q, a))
            story.append('')
        else:
            sent = tokenize(line)
            story.append(sent)
    return data


def get_stories(f, only_supporting=False, max_length=None):
    '''Given a file name, read the file, retrieve the stories,
    and then convert the sentences into a single story.

    If max_length is supplied,
    any stories longer than max_length tokens will be discarded.
    '''
    data = parse_stories(f.readlines(), only_supporting=only_supporting)
    flatten = lambda data: reduce(lambda x, y: x + y, data)
    data = [(flatten(story), q, answer) for story, q, answer in data
            if not max_length or len(flatten(story)) < max_length]
    return data


def vectorize_stories(data, word_idx, story_maxlen, query_maxlen):
    xs = []
    xqs = []
    ys = []
    for story, query, answer in data:
        x = [word_idx[w] for w in story]
        xq = [word_idx[w] for w in query]
        # let's not forget that index 0 is reserved
        y = np.zeros(len(word_idx) + 1)
        y[word_idx[answer]] = 1
        xs.append(x)
        xqs.append(xq)
        ys.append(y)
    return (pad_sequences(xs, maxlen=story_maxlen),
            pad_sequences(xqs, maxlen=query_maxlen), np.array(ys))
````

````
# Network Architecture

RNN = recurrent.LSTM
EMBED_HIDDEN_SIZE = 50
SENT_HIDDEN_SIZE = 100
QUERY_HIDDEN_SIZE = 100
BATCH_SIZE = 32
EPOCHS = 20
print('RNN / Embed / Sent / Query = \n {} / {} / {} / {}'.format(RNN,
                                                           EMBED_HIDDEN_SIZE,
                                                           SENT_HIDDEN_SIZE,
                                                           QUERY_HIDDEN_SIZE))
````

````
# Download Data
try:
    path = get_file('babi-tasks-v1-2.tar.gz',
                    origin='https://s3.amazonaws.com/text-datasets/'
                           'babi_tasks_1-20_v1-2.tar.gz')
except:
    print('Error downloading dataset, please download it manually:\n'
          '$ wget http://www.thespermwhale.com/jaseweston/babi/tasks_1-20_v1-2'
          '.tar.gz\n'
          '$ mv tasks_1-20_v1-2.tar.gz ~/.keras/datasets/babi-tasks-v1-2.tar.gz')
    raise
````

````
challenge_names = {
    'QA1': 'qa1_single-supporting-fact_{}.txt',
    'QA2': 'qa2_two-supporting-facts_{}.txt',
    'QA3': 'qa3_three-supporting-facts_{}.txt',
    'QA4': 'qa4_two-arg-relations_{}.txt',
    'QA5': 'qa5_three-arg-relations_{}.txt',
    'QA6': 'qa6_yes-no-questions_{}.txt',
    'QA7': 'qa7_counting_{}.txt',
    'QA8': 'qa8_lists-sets_{}.txt',
    'QA9': 'qa9_simple-negation_{}.txt',
    'QA10':'qa10_indefinite-knowledge_{}.txt',
    'QA11':'qa11_basic-coreference_{}.txt',
    'QA12':'qa12_conjunction_{}.txt',
    'QA13':'qa13_compound-coreference_{}.txt',
    'QA14':'qa14_time-reasoning_{}.txt',
    'QA15':'qa15_basic-deduction_{}.txt',
    'QA16':'qa16_basic-induction_{}.txt',
    'QA17':'qa17_positional-reasoning_{}.txt',
    'QA18':'qa18_size-reasoning_{}.txt',
    'QA19':'qa19_path-finding_{}.txt',
    'QA20':'qa20_agents-motivations_{}.txt'
}

def get_challenge_name(challenge_code, large=False):
  base = 'tasks_1-20_v1-2'
  size = 'en' if not large else 'en-10k'
  challenge = challenge_names[challenge_code]

  return os.path.join(base, size, challenge)
````

````
# Choose the challenge
# Challenge codes can be between QA1 - QA20

challenge_code = 'QA15'
large = False

challenge = get_challenge_name(challenge_code, large=large)
print('Challenge: ', challenge_code, ' -- ', challenge_names[challenge_code][:-7])
print(challenge)
````

````
# Process the stories into human readable format as well

def get_printable_stories(f, only_supporting=False, max_length=None):
    '''Given a file name, read the file, retrieve the stories.
    If max_length is supplied,
    any stories longer than max_length tokens will be discarded.
    '''
    data = parse_stories(f.readlines(), only_supporting=only_supporting)
    flatten = lambda data: reduce(lambda x, y: x + y, data)
    data = [(story, q, answer) for story, q, answer in data
            if not max_length or len(flatten(story)) < max_length]
    return data

def print_story(printable_story, include_answer=True):
    tx, qx, ax = printable_story
    print('\n'.join([' '.join(x) for x in tx]))
    print()
    print('Q:',' '.join(qx))

    if include_answer:
        print('A:', ax)


with tarfile.open(path) as tar:
    train_p = get_printable_stories(tar.extractfile(challenge.format('train')))
    test_p = get_printable_stories(tar.extractfile(challenge.format('test')))
````

````
# Print a random story from the training set

num = np.random.randint(0,len(train_p)+1)
print_story(train_p[num])
````

````
# Preprocess data for the model

with tarfile.open(path) as tar:
    train = get_stories(tar.extractfile(challenge.format('train')))
    test = get_stories(tar.extractfile(challenge.format('test')))

vocab = set()
for story, q, answer in train + test:
    vocab |= set(story + q + [answer])
vocab = sorted(vocab)

# Reserve 0 for masking via pad_sequences
vocab_size = len(vocab) + 1
word_idx = dict((c, i + 1) for i, c in enumerate(vocab))
story_maxlen = max(map(len, (x for x, _, _ in train + test)))
query_maxlen = max(map(len, (x for _, x, _ in train + test)))

x, xq, y = vectorize_stories(train, word_idx, story_maxlen, query_maxlen)
tx, txq, ty = vectorize_stories(test, word_idx, story_maxlen, query_maxlen)

print('vocab = {}'.format(vocab))
print('x.shape = {}'.format(x.shape))
print('xq.shape = {}'.format(xq.shape))
print('y.shape = {}'.format(y.shape))
print('story_maxlen, query_maxlen = {}, {}'.format(story_maxlen, query_maxlen))

print()
print('tx.shape = {}'.format(tx.shape))
print('txq.shape = {}'.format(txq.shape))
print('ty.shape = {}'.format(ty.shape))
````

````
# Build the model

print('Build model...')

sentence = layers.Input(shape=(story_maxlen,), dtype='int32')
encoded_sentence = layers.Embedding(vocab_size, EMBED_HIDDEN_SIZE)(sentence)
encoded_sentence = RNN(SENT_HIDDEN_SIZE)(encoded_sentence)

question = layers.Input(shape=(query_maxlen,), dtype='int32')
encoded_question = layers.Embedding(vocab_size, EMBED_HIDDEN_SIZE)(question)
encoded_question = RNN(QUERY_HIDDEN_SIZE)(encoded_question)

merged = layers.concatenate([encoded_sentence, encoded_question])
preds = layers.Dense(vocab_size, activation='softmax')(merged)

model = Model([sentence, question], preds)
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
````

````
# Train model

print('Training')
model.fit([x, xq], y,
          batch_size=BATCH_SIZE,
          epochs=EPOCHS,
          validation_split=0.05,
          verbose=1)
````

````
# Evaluate model

print('Evaluation')
loss, acc = model.evaluate([tx, txq], ty, batch_size=BATCH_SIZE)
print()
print('-'*80)
print('Test accuracy for {} task = {:.4f}'.format(challenge_names[challenge_code][:-7], acc))
````
